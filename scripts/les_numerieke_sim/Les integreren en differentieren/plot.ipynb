```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tesla = pd.read_csv('./race-sim-oop-csv/Tesla.csv')

print(tesla)
```

```python
bmw = pd.read_csv('./race-sim-oop-csv/BMW.csv')

print(bmw)
```

```python
t = tesla['time (s)']

x_tesla = tesla['position (m)']
v_tesla = tesla['speed (m/s)']

x_bmw = bmw['position (m)']
v_bmw = bmw['speed (m/s)']

fig, ax = plt.subplots()

ax.set_title("Racing results")

l1, = ax.plot(t, x_tesla, linewidth=1, color='red', linestyle='dashed', label='Tesla (x)')
l2, = ax.plot(t, v_tesla, 'r-', linewidth=1, label='Tesla (v)')

l3, = ax.plot(t, x_bmw, linewidth=1, color='blue', linestyle='dashed', label='BMW (x)')
l4, = ax.plot(t, v_bmw, 'b-', linewidth=1, label='BMW (v)')

ax.legend(handles=[l1, l2, l3, l4])

plt.show()
```
